```python
# Nome do arquivo - 'credit_card.csv'

import pandas as pd
df = pd.read_csv('credit_card.csv')
print(df.shape)
df.head(5)
```

```python
# Análise Exploratória: identificação de padrões nas variáveis explicativas segundo as distribuições
```

```python
import numpy as np
import matplotlib.pyplot as plt

for col in df.select_dtypes(include=[np.number]).columns:
    plt.figure()  # Cria uma nova figura para cada histograma
    plt.hist(df[col].dropna(), bins=30, edgecolor='k')  # Cria o histograma, removendo valores nulos
    plt.title(f'Histograma de {col}')  # Define o título do histograma
    plt.xlabel(col)  # Define o rótulo do eixo x
    plt.ylabel('Frequência')  # Define o rótulo do eixo y
    plt.grid(True)  # Adiciona uma grade ao gráfico
    plt.show()  # Mostra o histograma
```
